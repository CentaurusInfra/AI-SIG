```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('exp1.csv')
X = [128, 256, 512, 1024, 0]
alnair = {}
alnairlazy = {}
s3 = {}
for x in X:
    alnairlazy[str(x)] = data[data['MaxMemory(MB)']==x]['AlnairLazy'].values
    alnair[str(x)] = data[data['MaxMemory(MB)']==x]['Alnair'].values
    s3[str(x)] = data[data['MaxMemory(MB)']==x]['S3'].values

alnairlazy = pd.DataFrame(alnairlazy)
alnair = pd.DataFrame(alnair)
s3 = pd.DataFrame(s3)

x = ['128', '256', '512', '1024', 'inf']
plt.plot(x, alnairlazy.mean(), '-o', color='#7CFC00', label='AlnairLazy')
plt.fill_between(x, alnairlazy.mean()-alnairlazy.std(), alnairlazy.mean()+alnairlazy.std(), alpha=0.2,  color='#7CFC00', facecolor='#808080')
plt.plot(x, alnair.mean(), '-o', color='#CC4F1B', label='Alnair')
plt.fill_between(x, alnair.mean()-alnair.std(), alnair.mean()+alnair.std(), alpha=0.2,  color='#CC4F1B', facecolor='#FF9848')
plt.plot(x, s3.mean(), '-o', color='#1B2ACC', label='S3')
plt.fill_between(x, s3.mean()-s3.std(), s3.mean()+s3.std(), alpha=0.2, edgecolor='#1B2ACC', facecolor='#089FFF')

plt.legend(fontsize=12)
plt.grid()
plt.xlabel('MaxMemory(MB)', fontsize=16)
plt.ylabel('Inference Time(s)', fontsize=16)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.savefig('exp1.png', dpi=300)
```
